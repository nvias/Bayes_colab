#Bayesův klasifikátor
Právě jste otevřeli demo projekt, zaměřený na téma na Bayesův klasifikátor. 

Tento klasifikátor je z velké části postaven na znalostech lineární algebry. Ty zatím rozhodně nebudou nutné, pokud se ale budete chtít později řešeným problémům věnovat hlouběji, zjistíte, že znalosti tohoto předmětu jsou pro tuto problematiku nezbytné. 

Celý kód je napsán v Pythonu.

```
#import základních knihoven 
import matplotlib.pyplot as plt

#Náš dvoudimenzionální dataset – máme dvě části [x y]
dataset = [[3.393533211,2.331273381,0],
	[3.110073483,1.781539638,0],
	[1.343808831,3.368360954,0],
	[3.582294042,4.67917911,0],
	[2.280362439,2.866990263,0],
	[7.423436942,4.696522875,1],
	[5.745051997,3.533989803,1],
	[9.172168622,2.511101045,1],
	[7.792783481,3.424088941,1],
	[7.939820817,0.791637231,1]]

#vykreslení datasetu
for i in dataset:
  if (i[2]==0):
    plt.plot(i[0],i[1],'ro')
  else:
    plt.plot(i[0],i[1],'bo')
```

V tomto grafu můžete vidět vizualizaci datasetu v prostoru. Co si pod tím ale představit? 

Zvýrazněné body můžete vnímat jako zákazníky. Modré body představují zákazníky, kteří si koupili novou televizi, zatímco červené body představují zákazníky, kteří si koupili nový počítač. Předpokládejme, že osa X je věk zákazníka a osa Y je doba, kterou zákazník strávil na e-shopu. Díky tomu můžete vidět jasnou spojitost mezi věkem a dobou výběru. 

Co kdybychom teď přidali další parametr, například četnost návštěv?

Nyní zkusíme klasifikovat specifický bod [2,1].



```
from sklearn.naive_bayes import GaussianNB
#Numericko-matematická knihovna
import numpy as np
#Vytvoření modelu 
model = GaussianNB()

dataset_pc = [
  [3.393533211,2.331273381],
	[3.110073483,1.781539638],
	[1.343808831,3.368360954],
	[3.582294042,4.67917911],
	[2.280362439,2.866990263]]

np_dataset_pc = []
for i in dataset_pc:
  np_dataset_pc.append(np.array(i))

dataset_televize =  [
  [7.423436942,4.696522875],
	[5.745051997,3.533989803],
	[9.172168622,2.511101045],
	[7.792783481,3.424088941],
	[7.939820817,0.791637231]]

np_dataset_televize = []
for i in dataset_televize:
  np_dataset_televize.append(np.array(i))


dataset = [[3.393533211,2.331273381,0],
	[3.110073483,1.781539638,0],
	[1.343808831,3.368360954,0],
	[3.582294042,4.67917911,0],
	[2.280362439,2.866990263,0],
	[7.423436942,4.696522875,1],
	[5.745051997,3.533989803,1],
	[9.172168622,2.511101045,1],
	[7.792783481,3.424088941,1],
	[7.939820817,0.791637231,1]]

#Vykreslení datasetu
for i in dataset:
  if (i[2]==0):
    plt.plot(i[0],i[1],'ro')
  else:
    plt.plot(i[0],i[1],'bo')

dataset = [[3.393533211,2.331273381],
	[3.110073483,1.781539638],
	[1.343808831,3.368360954],
	[3.582294042,4.67917911],
	[2.280362439,2.866990263],
	[7.423436942,4.696522875],
	[5.745051997,3.533989803],
	[9.172168622,2.511101045],
	[7.792783481,3.424088941],
	[7.939820817,0.791637231]]

model.fit(dataset,['televize','televize','televize','televize','televize','Pc','Pc','Pc','Pc','Pc'])

print('Bod který klasifikujeme je zelený.')
plt.plot(2,1,'go')

point = np.array([[2,1]])

x = model.predict(point)

print('Kategorie:', x)
```

Vše, co potřebujete, je kód zobrazený výše, teď si jen řekneme, jak zadat vaše data do datasetu.Neprve musíte nahrát svá data, a poté v „model predict“ určíte, co se má zobrazit jako výsledek. :)
